# Crop Yield Predication

The aim of this notebook is to predict the crop yield using the data from the given dataset. The dataset has been taken from the Crop Yield data sheet.

![My Image](farm.jpeg)

### Data Dictionary

| Column Name     | Description                                  |
|-----------------|----------------------------------------------|
| Rain Fall (mm)  | Rainfall in millimeters                      |
| Temperature (C) | Temperature in Celsius                       |
| Fertilizer (kg) | Fertilizer in kilograms                      |
| Nitrogen (N)    | Nitrogen macro nutrient                      |
| Phosphorous (P) | Phosphorous macro nutrient                   |
| Potassium (K)   | Potassium macro nutrient                     |
| Yield (Q/acres) | Crop yield Quintals per acre                 |

```python
# Importing the modules
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Showing first five entries from dataset
df = pd.read_excel("crop yield data sheet.xlsx")
df.head()
```

# Data Preprocessing

```python
# checking the shape of the dataset
df.shape
```

```python
#Checking the data types of dataset
df.dtypes
```

Here the Temperature has object data type. we need to convert it to float data type. But first, I am checking the values in the column

```python
df['Temperatue'].unique()
```

The column has invalid value ":" .I need to remove this value before converting the column to float data type.

```python
#Dropping ":" from the Temperatue column
df = df[df['Temperatue'] != ':']
```

```python
#converting the Temperatue column to float
df['Temperatue'] = df['Temperatue'].astype(float)
```

```python
# Checking  for null values
df.isnull().sum()
```

Replacing missing values with median of the column

```python
# replacing missing values with median
columns = [df.columns]
for col in columns:
    df[col] = df[col].fillna(df[col].median())
```

Descriptive Statistics

```python
df.describe()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Yeild (Q/acre)', axis = 1), df['Yeild (Q/acre)'], test_size = 0.2, random_state = 42)
```

# Model Building

I will be using the following models to predict the crop yield:

1. Decision Tree Regressor
2. Random Forest Regressor

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

# Decision Tree Regressor object
dtr = DecisionTreeRegressor(random_state=42)
```

#### Hyperparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# Optional: clean NaNs from y_train
mask = ~np.isnan(y_train)
X_train_clean = X_train[mask]
y_train_clean = y_train[mask]

# Parameters for Grid Search
param_grid = {
    "max_depth": [2, 4, 6, 8],
    "min_samples_split": [2, 4, 6, 8],
    "min_samples_leaf": [2, 4, 6, 8],
    "random_state": [0, 42]  # Optional  some models ignore this during search
}

# Grid Search object
grid_search = GridSearchCV(
    estimator=dtr,
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    verbose=1
)

# Fit the Grid Search
grid_search.fit(X_train_clean, y_train_clean)

# Best parameters
print("Best Hyperparameters:", grid_search.best_params_)
```

```python
# Get best hyperparameters, excluding 'random_state' if it exists
best_params = grid_search.best_params_.copy()  # Make a copy to avoid modifying the original
best_params.pop("random_state", None)  # Remove 'random_state' if present

# Train Decision Tree Regressor with best parameters
dtr = DecisionTreeRegressor(**best_params, random_state=42)  # Manually set random_state

# Fitting the model
dtr.fit(X_train, y_train)

# Training score
print("Training Score:", dtr.score(X_train, y_train))

# Prediction
d_pred = dtr.predict(X_test)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

#Random Forest Regressor object
rfr = RandomForestRegressor()
```

#### Hyperparameter Tuning using RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

# Parameters for Randomized Search
param_dist = {
    "n_estimators": [100, 200, 300],
    "max_depth": [2, 4, 6, 8],
    "min_samples_split": [2, 4, 6, 8],
    "min_samples_leaf": [2, 4, 6, 8],
    "random_state": [0, 42]
}

# Randomized Search object
random_search = RandomizedSearchCV(
    estimator=rfr,
    param_distributions=param_dist,
    n_iter=10,  # Reduced trials
    cv=3,  # Reduced cross-validation folds
    n_jobs=-1,  # Use all CPU cores
    verbose=1,
    random_state=42
)

# Fitting the Randomized Search object
random_search.fit(X_train, y_train)

# Best parameters
print("Best Hyperparameters:", random_search.best_params_)
```

```python
# Train model with best parameters
rfr_best = RandomForestRegressor(**random_search.best_params_)

# Fitting the model
rfr_best.fit(X_train, y_train)

# Training score
print("Training Score:", rfr_best.score(X_train, y_train))

# Predictions
r_pred = rfr_best.predict(X_test)
```

### Model Evaluation

#### Distribution Plot

```python
fig, ax = plt.subplots(1,2,figsize=(15, 5))

#plotting the actual vs predicted values for decision tree regressor
sns.distplot(y_test, ax = ax[0]).set_title('Decision Tree Regressor')
sns.distplot(d_pred, ax = ax[0])

#plotting the actual vs predicted values for random forest regressor
sns.distplot(y_test, ax = ax[1]).set_title('Random Forest Regressor')
sns.distplot(r_pred, ax = ax[1])
```

### Evaluation Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("Decision Tree Regressor")
print("Mean Squared Error: ", mean_squared_error(y_test, d_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, d_pred))
print("R2 Score: ", r2_score(y_test, d_pred))
print("\n")
print("Random Forest Regressor")
print("Mean Squared Error: ", mean_squared_error(y_test, r_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, r_pred))
print("R2 Score: ", r2_score(y_test, r_pred))
```

### Feature Importance

```python
# Ensure RandomForestRegressor is trained
rfr = RandomForestRegressor(n_estimators=100, random_state=42)
rfr.fit(X_train, y_train)  #  Fit the model

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
fig.subplots_adjust(wspace=0.5)

# Decision Tree Regressor
feature_df_dt = pd.DataFrame({'Features': X_train.columns, 'Importance': dtr.feature_importances_})
feature_df_dt.sort_values(by='Importance', ascending=False, inplace=True)
sns.barplot(x='Importance', y='Features', data=feature_df_dt, ax=ax[0])
ax[0].set_title('Decision Tree Regressor')

# Random Forest Regressor
feature_df_rf = pd.DataFrame({'Features': X_train.columns, 'Importance': rfr.feature_importances_})
feature_df_rf.sort_values(by='Importance', ascending=False, inplace=True)
sns.barplot(x='Importance', y='Features', data=feature_df_rf, ax=ax[1])
ax[1].set_title('Random Forest Regressor')

plt.show()
```
